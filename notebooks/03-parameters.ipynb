In this notebook we will learn about parameters.

````python
# Numerical arrays.
import numpy as np

# Plots.
import matplotlib.pyplot as plt
````

````python
# A range of x values.
x = np.linspace(0.0, 10.0, 20)
x
````

````python
# A range of y values based on x.
y = 4.0 * x - 2.0 + np.random.normal(0.0, 1.0, len(x))
y
````

````python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot x vs y.
ax.plot(x, y, 'k.');
````

````python
# Use x and y to try to find the original parameters.
params = np.polyfit(x, y, 1)
params
````

````python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot x vs y.
ax.plot(x, y, 'k.')

# Plot x vs parameters on x.
ax.plot(x, params[0] * x + params[1], 'b-');
````

````python
# Use x and y to try to find the original parameters.
params = np.polyfit(x, y, 2)
params
````

````python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot x vs y.
ax.plot(x, y, 'k.')

xtmp = np.linspace(-1000.0, 100.0, 10000)

# Plot x vs parameters on x.
ax.plot(xtmp, params[0] * xtmp**2 + params[1] * xtmp + params[2], 'b-');
````

````python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Temporary x values.
xtmp = np.linspace(-9.0, 11.0, 1000)

# Plot x vs parameters on x.
ax.plot(xtmp, xtmp**2 - 2.0 * xtmp + 1.0, 'b-');
````

````python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Temporary x values.
xtmp = np.linspace(-9.0, 11.0, 1000)

# Plot x vs parameters on x.
ax.plot(xtmp, xtmp**3 - 4.0 * xtmp**2 - 2.0 * xtmp + 1.0, 'b-');
````

````python
# Use x and y to try to find the original parameters.
params = np.polyfit(x, y, 3)
params
````

````python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot x vs y.
ax.plot(x, y, 'k.')

# Plot x vs parameters on x.
ax.plot(x, np.polyval(params, x), 'b-');
````

````python
# Highest power of x in the polynomial.
degree = 100

# Use x and y to try to find the original parameters.
params = np.polyfit(x, y, degree)

# Show the parameters.
print(params)

# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot x vs y.
ax.plot(x, y, 'k.')

# Plot x vs parameters on x.
ax.plot(x, np.polyval(params, x), 'b-');
````
